```python
from src.training       import create_variables,training_model_rf, variables_scaler
from src.data_loader    import create_df
from src.evaluation     import show_confusion_matrix, show_metricas, show_feature_importances, show_auc_roc, show_especificidad, get_feature_importances
from sklearn.linear_model import LogisticRegression

df_model_rl = create_df('clean_data')
X_train, X_test, y_train, y_test = create_variables(df_model_rl,False)

modelo = LogisticRegression(class_weight='balanced')

X_train_scaled, X_test_scaled = variables_scaler(X_train, X_test)

modelo.fit(X_train_scaled, y_train)
pred = modelo.predict(X_test_scaled)

metricas, acc = show_metricas(y_test, pred)
print("\nReporte de Clasificación del Modelo Final:")
print(metricas)
```
